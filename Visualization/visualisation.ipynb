# Ensemble Results

```python
PATH_ENSEMBLE = "/Users/sfoell/Library/CloudStorage/OneDrive-ETHZurich/ETH-GDBD/Publications/2022_ICML_GDU4DG/Results/Digits5/Ensemble"
results = sorted(glob.glob(PATH_ENSEMBLE + '/**/spec_*.csv', recursive = True))

data = pd.DataFrame()
out = pd.DataFrame()

for result in results:
    data = data.append(pd.read_csv(result, index_col = 'run_id'))

for target in data.target_domain.unique():
    condition = (data.target_domain == target)
    mean = round(data.loc[condition].categorical_accuracy.mean()*100, 2)
    std = round(data.loc[condition].categorical_accuracy.std()*100, 2)
    out.at[0, str(target).upper()] = str(mean)+" (" + str(std)+")"

#out.to_excel(PATH_ENSEMBLE + '/ensmble_results.xlsx')
print(out.to_markdown())
```

```python
plt.rcParams.update({'font.size': 11})
specs = {"mnist": ["MNIST", (0.98, 0.99)],
        "mnistm": ["MNIST-M", (0.45, 0.75)],
        "svhn": ["SVHN", (0.65, 0.85)],
        "syn": ["SYN", (0.83, 0.9)],
        "usps": ["USPS", (0.94, 0.98)],
        }


for target_domain in ["mnist", "mnistm", "syn", "svhn", "usps"]:
    mmd_matrixs = glob.glob("/Users/sfoell/Library/CloudStorage/OneDrive-ETHZurich/ETH-GDBD/Publications/2022_ICML_GDU4DG/Results/Digits5/Visualization/vis_elementary_domain_bases/run_0/"+target_domain+"/**/mmd_matrix_*.csv")
    fig, ax = plt.subplots(3,2, figsize=(20,30))
    fig.suptitle(specs[target_domain][0], size= 12)
    #fig.tight_layout()
    ax = np.ravel(ax)
    i = 0
    print(target_domain)
    for file in mmd_matrixs:
        similarity = file.split("_")[-1].split(".")[0]
        if similarity == "SIMILARITY":
            similarity_name = "CS"
        else:
            similarity_name = similarity
        mmd_data = pd.read_csv(file, index_col="Unnamed: 0")
        mmd_data.rename(axis=0, mapper={"V_0": r'$V_{0}$',
                                                          "V_1": r'$V_{1}$',
                                                         "V_2": r'$V_{2}$',
                                                         "V_3": r'$V_{3}$',
                                                        "V_4": r'$V_{4}$'
                                                         }, inplace = True)
        mmd_data.rename(axis=1, mapper={"V_0": r'$V_{0}$',
                                                          "V_1": r'$V_{1}$',
                                                         "V_2": r'$V_{2}$',
                                                         "V_3": r'$V_{3}$',
                                                        "V_4": r'$V_{4}$'
                                                         }, inplace = True)
        matrix = np.triu(mmd_data)
        sns.heatmap(mmd_data, annot=True, linewidths=.8, cbar_kws={"shrink": .80},
                    ax = ax[i],  mask=matrix)
        ax[i].set_title("MMD heatmap for: %s" %similarity_name)
        
        x_data = glob.glob("/Users/sfoell/Library/CloudStorage/OneDrive-ETHZurich/ETH-GDBD/Publications/2022_ICML_GDU4DG/Results/Digits5/Visualization/vis_elementary_domain_bases/run_0/"+target_domain+"/**/x_data_df_*"+similarity+".csv")
        x_data = pd.read_csv(x_data[0])
        
        domain_basis = glob.glob("/Users/sfoell/Library/CloudStorage/OneDrive-ETHZurich/ETH-GDBD/Publications/2022_ICML_GDU4DG/Results/Digits5/Visualization/vis_elementary_domain_bases/run_0/"+target_domain+"/**/domain_basis_df_*"+similarity+".csv")
        domain_basis = pd.read_csv(domain_basis[0])
        
        embedding_x_data_array = TSNE(n_components=2, perplexity=5, n_iter=500, random_state=1234).fit_transform(x_data.iloc[:, :-2])
        embedded_domain_basis_array= TSNE(n_components=2, perplexity=5, n_iter=500, random_state=1234).fit_transform(domain_basis.iloc[:, :-2])

        embedding = np.concatenate((embedding_x_data_array,embedded_domain_basis_array ))

        index = np.concatenate((x_data.iloc[:,-1], domain_basis.iloc[:,-1])).flatten()
        replacements = {'V_0':r'$V_{0}$',
               'V_1':r'$V_{1}$',
               'V_2':r'$V_{2}$',
               'V_3':r'$V_{3}$',
               'V_4':r'$V_{4}$'}
        replacer = replacements.get  # For faster gets.

        index = [replacer(n, n) for n in index]
        
        
        i+=1
        sns.scatterplot(x=embedding[:, 0], y=embedding[:, 1],hue = index, palette="bright", ax = ax[i])
        ax[i].set_title("t-SNE embedding for: %s" %similarity)
        i+=1
        
        
    plt.savefig("/Users/sfoell/Desktop/"+target_domain.upper()+".eps", format="eps", bbox_inches= "tight")
```
